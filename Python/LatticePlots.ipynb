```python
import matplotlib.pyplot as plt
import numpy as np
import itertools
```

```python
plt.rcParams['figure.figsize']=(10,8)
plt.style.use('seaborn-whitegrid')
```

```python
L = []
for i in range(1,4):
    L.append(np.tile(['u','d'], i))
```

```python
def CRRLattice(T):

    for i in range(T):
        x = [1, 0, 1]
        for j in range(i):
            x.append(0)
            x.append(1)
        x = np.array(x) + i
        y = np.arange(-(i+1), i+2)[::-1]
        plt.plot(x, y, 'o-', color='darkred')

        p=[''.join(x) for x in itertools.product('ud',  repeat=i+1)]
        y = list(y)
        y.remove(0)
        z = np.array(y)
        cord = np.where(z > 0, z-0.3, z+0.3)
        for k in range(len(y)):
            plt.annotate(p[k], (i+0.5, cord[k]),fontweight='bold',color='darkblue')
```

```python
CRRLattice(3)
#plt.savefig('./lattice.pdf',format='pdf', dpi=400)
```
